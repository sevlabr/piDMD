### Circulant (+ TLS, unitary, symmetric, skew-symmetric)

#### Simple circulant

```python
from piDMD.gens import gen_circulant
```

### Symmetric / Skew-symmetric

```python
# set random seed
np.random.seed(1)

# generate symmetric (Hermitian) model
n_ = 20
trueA = np.random.randn(n_, n_)
trueA = (trueA + trueA.T) / 2
trueVals, _ = LA.eig(trueA)

piA, piVals, piVecs, exA, exVals, exVecs = simple_test(
    n=n_, m=1000, trueA=trueA, method="symmetric"
)
```

### Tridiagonal and symmetric

#### Variable diagonal width

```python
n_ = 10

diag = 2*np.ones((n_, 2))
diag[3, 0] = 1
diag[3, 1] = 5
diag[5, 0] = 5
diag[6, 1] = 4
print(f"Diagonal shape defines as:\n{diag}\n")

# set random seed
np.random.seed(1)

# generate variable-diagonal (local) model
diag_1 = np.random.randn(n_-1)
diag_2 = np.random.randn(n_)
diag_3 = np.random.randn(n_-1)
trueA = np.diag(diag_1, k=-1) + np.diag(diag_2, k=0) + np.diag(diag_3, k=1)
trueA[3, 2] = 0
trueA[3, 5:8] = np.random.randn(3)
trueA[5, 1:4] = np.random.randn(3)
trueA[6, 8:10] = np.random.randn(2)
trueVals, _ = LA.eig(trueA)

# tls means total least squares
piA, piVals, piVecs, exA, exVals, exVecs = simple_test(
    n=n_, m=1000, trueA=trueA, method="diagonaltls", diag=diag
)
```

```python
plot_eigenvalues(trueVals, piVals, exVals, pi_title="piDMD variable-diagonaltls")
```

```python
I = np.eye(n_)
plot_matrices(trueA, piA(I), exA(I))
```
